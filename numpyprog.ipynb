random([size, dtype, out])
	

Return random floats in the half-open interval [0.0, 1.0).

https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.random.html#numpy.random.Generator.random

[ is inclusive, ) is exclusive, tuple = array

ndarray = number of dimensions in the array

```python
import numpy as np
```

```python
rng = np.random.default_rng()

rng.random(5)
```

```python
rng.random(5,4)
```

```python
rng.random((5,4)) # encapsulate that
```

```python
rng.random((5,4,3,7,6)) # add extra dimensions
```

```python
rng.random((5,4)) # encapsulate that
```

```python
rng = np.random.default_rng()

x = rng.random((5,4))
```

```python
print (x)
```

```python
x = rng.random(10000)
print (x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x)
plt.show()
```

integers(low[, high, size, dtype, endpoint])
	

Return random integers from low (inclusive) to high (exclusive), or if endpoint=True, low (inclusive) to high (inclusive).

https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.integers.html#numpy.random.Generator.integers

```python
rng = np.random.default_rng()

rng.integers(2, size=10)
```

```python
rng.integers(1, size=10)
```

```python
rng.integers(5, size=(2, 4))
```

```python
rng.integers(1, [3, 5, 10])
```

```python
rng.integers([1, 5, 7], 10)
```

```python
rng.integers([1, 3, 5, 7], [[10], [20]], dtype=np.uint8)
```
