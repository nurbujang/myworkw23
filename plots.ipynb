$ f(x) = x^2 -4x + 4 $

```python
# Code the above function.
def f(x):
  return x**2 - (4.0 * x) + 4.0
```

```python
# numpy array of x values with problems

x = np.array([50.0, 60.0, 70.0, 100.0, 105.0, 123.45])
```

```python
y = f(x)
y
```

```python
# create an empty plot
fig, ax = plt.subplots()

# plot x vs y
ax.plot(x, y); # plot not nice
```

```python
# try

x = np.array([-3.0, -2, -1, 0, 1, 2, 3, 4])
y = f(x)
y

# create an empty plot
fig, ax = plt.subplots()

# plot x vs y
ax.plot(x, y); 
```

```python
# try

x = np.array([-2.0, -1, 0, 1, 2, 3, 4, 5, 6]) 
# annoying to write these values, so use arange, linspace to set x values

y = f(x)
y

# create an empty plot
fig, ax = plt.subplots()

# plot x vs y
ax.plot(x, y); 
```

```python
# try

x = np.array([-2.0, -1, 0, 1, 2, 3, 4, 5, 6]) 
# annoying to write these values, so use arange, linspace to set x values

y = f(x)
y

# create an empty plot
fig, ax = plt.subplots()

# more common to see scatter plot on top of the line to better see
# plot x vs y, interpolate
ax.plot(x, y); 

# plot x vs y, scatterplot
ax.plot(x, y, '.'); 

```

```python
# how to fix interpolation?

# add more data points, linspace, arange

# Array of x values:

x = np.arange(-1, 5, 0.01) # low, high, add 0.01 each time
y = f(x)
y

# create an empty plot
fig, ax = plt.subplots()

# plot x vs y, interpolate
ax.plot(x, y)

# plot x vs y, scatterplot
ax.plot(x, y, '.'); 
```

```python
# OR LINSPACE (MUCH PREFERRED)
# Numpy Array of x values.
x = np.linspace(-1.0, 5.0, 601) # LOW, HIGH, HOW many numbers u want between them, 601-1=600
```

```python
# Calculate the y values.
y = f(x)
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Interpolate.
ax.plot(x, y)
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-4.0, 4.0, 801)

# Plot x^2.
ax.plot(x, x**2);
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-4.0, 4.0, 801)

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2 + 1.0.
ax.plot(x, x**2 + 1.0);
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-4.0, 4.0, 801)

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2 + 1.0.
ax.plot(x, x**2 + 1.0, label='$x^2 + 1$')

# Plot x^2 - 1.0.
ax.plot(x, x**2 - 1.0, label='$x^2 - 1$')

# Plot x^2 - 3.0.
ax.plot(x, x**2 - 3.0, label='$x^2 - 3$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-12.0, 12.0, 801)

# Plot x^2.
ax.plot(x, x**2, label='$x^2$')

# Plot x^2 + x.
ax.plot(x, x**2 + x, label='$x^2 + x$')

# Plot x^2 - x.
ax.plot(x, x**2 - x, label='$x^2 - x$')

# Plot x^2 - 4.0x.
ax.plot(x, x**2 - (4.0 * x), label='$x^2 - 4x$')

# Turn on legend.
ax.legend();
```

```python
# Create an empty plot.
fig, ax = plt.subplots()

# Range of x values.
x = np.linspace(-12, 12, 801)

# Plot log (natural).
ax.plot(x, np.log(x), label='$\ln(x)$')

# Plot log base 2.
ax.plot(x, np.log2(x), label='$\log_2(x)$')

# Plot log base 10.
ax.plot(x, np.log10(x), label='$\log_{10}(x)$')

# Turn on legend.
ax.legend();
# error bc it doesmt like divide by 0
```
