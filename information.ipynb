In probability theory and statistics, the binomial distribution with parameters n and p is the discrete probability distribution of the number of successes in a sequence of n independent experiments, each asking a yesno question, and each with its own Boolean-valued outcome: success (with probability p) or failure (with probability q = 1  p {\displaystyle q=1-p}). A single success/failure experiment is also called a Bernoulli trial or Bernoulli experiment, and a sequence of outcomes is called a Bernoulli process; for a single trial, i.e., n = 1, the binomial distribution is a Bernoulli distribution.

```python
# The Bernoulli distribution is just the binomial with the first parameter as 1.
def bernoulli(p, size=None):
  return np.random.binomial(1, p, size) # 1 = number of trials u do, size = number of times ure doing it

# FLIP 1 COIN, BUT FLIP 100 TIMES VS FLIP 100 COINS, AND FLIP 1 TIME
```

```python
# Flip one hundred fair coins.
bernoulli(0.5, 100)
```

```python
# Flip one hundred unfair coins.
bernoulli(0.7, 100)
```

```python
# Roll two dice.
# generate 200,000 integers between 1 and 7
# but group them up so it's a 2D array, outer barray will have 100000 thing in it, the inside array will hv 2 in it
# so 100000 eoqs, 2 columns
double_rolls = rng.integers(1, 7, (100000, 2))

# Show a few.
double_rolls
```

```python
# Sum the two dice. (how many different ways to get sum, say 11?)
sum_double_rolls = double_rolls.sum(axis=1) # axis 0 is outer array(100000), axis 1 is inner array [5,6] or [6,5]
# if axis=0, it will sum down the column containing 100000 things
# axis = 1 will add in row

# Show a few.
sum_double_rolls
```

```python
# Count the unique values.
x, counts = np.unique(sum_double_rolls, return_counts=True)
```

```python
# Create an empty plot.
fig, ax = plt.subplots(figsize=(5, 3))

# Plot the bar chart.
ax.bar(x, counts)

# Make the x axis look sensible.
ax.set_xticks(x);
```

## Information Entropy

https://en.wikipedia.org/wiki/Entropy_(information_theory)

https://www.quantamagazine.org/how-claude-shannons-concept-of-entropy-quantifies-information-20220906/

https://mathshistory.st-andrews.ac.uk/Biographies/Shannon/

Entropy measures the expected (i.e., average) amount of information conveyed by identifying the outcome of a random trial.[5]:67 This implies that rolling a die has higher entropy than tossing a coin because each outcome of a die toss has smaller probability ( p = 1 / 6 {\displaystyle p=1/6}) than each outcome of a coin toss ( p = 1 / 2 p=1/2).

Consider a coin with probability p of landing on heads and probability 1  p of landing on tails. The maximum surprise is when p = 1/2, for which one outcome is not expected over the other. In this case a coin flip has an entropy of one bit. (Similarly, one trit with equiprobable values contains log 2  3 \log _{2}3 (about 1.58496) bits of information because it can have one of three values.) The minimum surprise is when p = 0 or p = 1, when the event outcome is known ahead of time, and the entropy is zero bits. When the entropy is zero bits, this is sometimes referred to as unity, where there is no uncertainty at all  no freedom of choice  no information. Other values of p give entropies between zero and one bits.

Information theory is useful to calculate the smallest amount of information required to convey a message, as in data compression. For example, consider the transmission of sequences comprising the 4 characters 'A', 'B', 'C', and 'D' over a binary channel. If all 4 letters are equally likely (25%), one can not do better than using two bits to encode each letter. 'A' might code as '00', 'B' as '01', 'C' as '10', and 'D' as '11'. However, if the probabilities of each letter are unequal, say 'A' occurs with 70% probability, 'B' with 26%, and 'C' and 'D' with 2% each, one could assign variable length codes. In this case, 'A' would be coded as '0', 'B' as '10', 'C' as '110', and D as '111'. With this representation, 70% of the time only one bit needs to be sent, 26% of the time two bits, and only 4% of the time 3 bits. On average, fewer than 2 bits are required since the entropy is lower (owing to the high prevalence of 'A' followed by 'B'  together 96% of characters). The calculation of the sum of probability-weighted log probabilities measures and captures this effect. English text, treated as a string of characters, has fairly low entropy; i.e. it is fairly predictable. We can be fairly certain that, for example, 'e' will be far more common than 'z', that the combination 'qu' will be much more common than any other combination with a 'q' in it, and that the combination 'th' will be more common than 'z', 'q', or 'qu'. After the first few letters one can often guess the rest of the word. English text has between 0.6 and 1.3 bits of entropy per character of the message.

$ H(X) := - \sum_{x \in X} p(x) \log p(x) $

**For a bit**

$ H(X) = -p(0) \log p(0) - p(1) \log p(1) $

$ H(X) = -(1 - p(1)) \log (1 - p(1)) - p(1) \log p(1) $

Let $p = p(1)$.

$ H(p) = -(1 - p) \log (1 - p) - p \log p $

```python
# Entropy as above.
def H(p):
  return -(1 - p) * np.log2(1.0 - p) - p * np.log2(p)
```

```python
# Entropy of 0.5.
H(0.5)
```

```python
# Create an empty plot.
fig, ax = plt.subplots(figsize=(12,6))

# p is a probability.
p = np.linspace(0.00000001, 0.99999999, 10001) # 0.00000001 bc u cant put 0,  0.99999999 bc u cant put 1

# Plot H(p).
ax.plot(p, H(p));

# Shannon's entropy formula for flipping a coin, same for communication table
# if u have a fair 50/50 and flip, when u see the outcome, u learn 1 bit of info (0 or 1)
# but if a coin is biased, eg prob of heads is 0.4, when u see outcome, u dont quite get as much info (u dont get 1)
# prob of seeing tail (0.6) is outweighing prob of seeing heads (0.4)
```
