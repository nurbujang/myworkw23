```python
# store it in pandas dataframe

df = pd.read_csv("https://datahub.io/core/co2-ppm/r/co2-mm-mlo.csv")
```

```python
# index out Date column
# will output dtype:object (still generic. Not yet recognized as date time)

df["Date"]
```

```python
# convert into date time
# will output dtype:datetime64
pd.to_datetime(df["Date"])
```

```python
df['datetime'] = pd.to_datetime(df["Date"])
```

```python
df # will output new column datetime
```

```python
import seaborn as sns

# https://seaborn.pydata.org/tutorial/relational.html

# https://seaborn.pydata.org/generated/seaborn.lineplot.html
```

```python
sns.lineplot(data=df, x="datetime", y="Interpolated")
sns.lineplot(data=df, x="datetime", y="Trend")
```

```python
# lets say we only want to do first 1000 rows
# https://stackoverflow.com/questions/15008970/way-to-read-first-few-lines-for-pandas-dataframe

df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv", skiprows=23, nrows=1000, low_memory=False)
```
